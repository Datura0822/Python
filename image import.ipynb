```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=np.array(Image.open('d:/bs/zzh1.jpg'))  #打开图像并转化为数字矩阵
plt.figure("dog")
plt.imshow(img)
plt.axis('on')
plt.show()
```

```python
img=np.array(Image.open('d:/bs/zzh1.jpg').convert('L'))
print(img)
rows,cols=img.shape

for i in range(rows):
    for j in range(cols):
        if (img[i,j]<=80):
            img[i,j]=0
        else:
            img[i,j]=1
print(img)         
plt.figure("lena")
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.show()
```

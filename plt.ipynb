```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.linspace(0, 4)
plt.plot(theta, 0 * np.cos(theta) + 0 * np.sin(theta), label = 'r = 0*cos(theta)+0*sin(theta)')
plt.plot(theta, 2 * np.cos(theta) + 1 * np.sin(theta), label = 'r = 2*cos(theta)+1*sin(theta)')
plt.plot(theta, 4 * np.cos(theta) + 2 * np.sin(theta), label = 'r = 4*cos(theta)+2*sin(theta)')

plt.xlabel('theta label')
plt.ylabel('r label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```
